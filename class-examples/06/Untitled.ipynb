```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

def make_student_success(n=500, seed=1):
    rng = np.random.default_rng(seed)
    hours = rng.integers(0, 12, n)
    attendance = rng.integers(40, 100, n)
    sleep = rng.integers(4, 9, n)
    practice_tests = rng.integers(0, 6, n)
    y = (((hours >= 5) & (attendance >= 70) & (sleep >= 6)) | (practice_tests >= 4)).astype(int)
    y = np.where(rng.random(n) < 0.07, 1-y, y)  # a bit of noise
    X = pd.DataFrame({"hours": hours, "attendance": attendance, "sleep": sleep, "practice_tests": practice_tests})
    return X, y


X, y = make_student_success()

#print(X)
#print(y)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = KNeighborsClassifier(n_neighbors=5)

model.fit(X_train, y_train)
pred = model.predict(X_test)


print("Accuracy:", accuracy_score(y_test, pred))




for k in [1,3,5,7,9]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print("k:",k,"accuracy:",round(acc,3))


```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np


def make_healthcare(n=600, seed=4):
    rng = np.random.default_rng(seed)
    age = rng.integers(1, 90, n)
    fever = rng.normal(37.0, 1.0, n)
    oxygen = rng.normal(97, 2.5, n)
    high_risk = (((age > 60) & (fever > 38.0)) | (oxygen < 94)).astype(int)
    high_risk = np.where(rng.random(n) < 0.06, 1-high_risk, high_risk)
    X = pd.DataFrame({"age": age, "fever": fever, "oxygen": oxygen})
    y = high_risk
    return X, y

X, y = make_healthcare()


#print(X)
#print(y)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = KNeighborsClassifier(5)

model.fit(X_train,y_train)

print("Accuracy:", model.score(X_test,y_test))




for k in [1,3,5,7,9]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print("k:",k,"accuracy:",round(acc,3))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np


def make_fraud(n=1200, seed=5):
    rng = np.random.default_rng(seed)
    amount = rng.normal(60, 25, n).clip(1, None)
    is_international = rng.integers(0, 2, n)
    night = rng.integers(0, 2, n)
    fraud = ((amount > 95) & (is_international == 1) & (night == 1)).astype(int)
    # make fraud rare
    fraud = np.where(rng.random(n) < 0.93, 0, fraud)
    X = pd.DataFrame({"amount": amount, "is_international": is_international, "night": night})
    y = fraud
    return X, y

X, y = make_fraud()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = KNeighborsClassifier(5)

model.fit(X_train,y_train)

print("Accuracy:", model.score(X_test,y_test))
```
